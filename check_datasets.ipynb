```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.load("datasets/mask.npz")["img"]
y = np.load("datasets/face.npz")["img"]
```

```python
for n in [1, 100, 500, 1000, 5000, 10000]:
    ax = plt.subplot(1, 2, 1)
    ax.imshow(x[n])
    ax.axis("off")
    ax = plt.subplot(1, 2, 2)
    ax.imshow(y[n])
    ax.axis("off")
    plt.show()
```
